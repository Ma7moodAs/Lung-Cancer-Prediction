## Dataset Loading and inspection

```python
df = pd.read_csv(r'C:\Users\User\Desktop\Lung Cancer Prediction\data\survey lung cancer.csv')
print('DataFrame has been loaded successfully.')
display(df.head())
print(f'The shape of the DataFrame is{df.shape}')
print(f'The DataFrame has {df.shape[0]} rows, and {df.shape[1]} columns')
print('\nDataFrame Information:')
print(df.info())
print('\nStatistical Summary:')
display(df.describe(include='all').T)
columns = df.columns.to_list()
print('\nColumn Names:')
print(columns)
```

```python
 # Checking for the missing values and duplicated values
print(f'Number of Missing Values: {df.isna().sum().sum()}')
print(f'Number of Duplicated Values: {df.duplicated().sum()}')

duplaicated_values = df[df.duplicated(keep=False)]
print(duplaicated_values.shape)
print('\nDuplicated Values:')
display(duplaicated_values)
print('\nStatistics for Duplicated Valued:')
display(duplaicated_values.describe(include='all').T)
```

## EDA

```python
custom_palette = ['#f1a0a1', '#4c7cba']
plt.figure(figsize=(8,6),dpi=100)
bars = sns.countplot(x='LUNG_CANCER',
              data=df,
              palette=custom_palette)
for bar in bars.patches:
  y_value = bar.get_height()
  x_pos = bar.get_x() + (bar.get_width()/2)
  if y_value > 0:
    plt.text(x_pos,y_value,int(y_value),ha='center',va='bottom')
  bar.set_hatch('/')
  bar.set_edgecolor('white')
plt.title('Disribution of Lung Cancer',fontdict={'fontsize':12,'fontweight':'bold'})
plt.xlabel('Lung Cancer')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

#Creation of a subplots

fig,axes = plt.subplots(1,2,figsize=(12,6))
sns.countplot(x='LUNG_CANCER',
              data=df,
              ax=axes[0],
              palette=custom_palette)
axes[0].set_title('Count of Lung Cancer Cases',fontdict={'fontsize':12})
axes[0].set_xlabel('Lung Cancer')
axes[0].set_ylabel('Count')
for bar in axes[0].patches:
  y_val = bar.get_height()
  x_pos = bar.get_x() + (bar.get_width()/2)
  if y_val > 0:
    axes[0].text(x_pos,y_val,int(y_val),ha='center',va='bottom')
  bar.set_edgecolor('white')
  bar.set_hatch('/')
axes[0].set_axisbelow(True)
axes[0].grid(axis='y',linestyle='--',linewidth=0.7,color='gray')
sns.despine(left=True,bottom=False)

lung_cancer_counts = df['LUNG_CANCER'].value_counts()
axes[1].pie(lung_cancer_counts,
            labels=lung_cancer_counts.index,
            autopct='%1.2f%%',
            startangle=90,
            colors=custom_palette)
axes[1].set_title('Proportion of Lung cancer',fontdict={'fontsize':12})

center_circle =  plt.Circle((0,0),radius=0.8,fc='white')
fig = plt.gcf()
fig.gca().add_artist(center_circle)
#axes[1].axis('equal')
plt.tight_layout()
plt.savefig('Proportion of Lung Cancer.png')
plt.show()
```

```python
sns.histplot(x='AGE',data=df,kde=True,color='#eda4a5')
plt.title('Age Distribution',fontdict={'fontsize':12})
plt.xlabel('Age')
plt.ylabel('Frequency')
sns.despine()
plt.grid(axis='y',linestyle='--',linewidth=0.7,color='gray')
```

#### The distribution of the age is left skewed, so we will apply a variable transformation on it to get the distribution a bit gaussian

```python

df['transformed_age'] = np.sqrt(df['AGE'])
sns.histplot(x='transformed_age',data=df,kde=True,color='#eda4a5')
plt.title('Transformed Age Distribution',fontdict={'fontsize':12})
plt.xlabel('Age')
plt.ylabel('Frequency')
sns.despine()
plt.grid(axis='y',linestyle='--',linewidth=0.7,color='gray')
plt.tight_layout()
plt.show()
```

```python
# Distribution of categorical columns
def Distribution_of_Cat_cols(col_name,data=df):
  # Count Plots for the total number of the distribution
  fig,axes = plt.subplots(1,2,figsize=(10,5))
  sns.countplot(x=col_name,
                data=data,
                palette=['#7ecfed','#eda17e'],
                ax=axes[0])
  axes[0].set_title(f'Distribution of {col_name}',fontdict={'fontsize':12})
  axes[0].set_xlabel(col_name)
  axes[0].set_ylabel('Frequency')
  for bar in axes[0].patches:
    y_val = bar.get_height()
    x_pos = bar.get_x() + (bar.get_width()/2)
    if y_val > 0:
      axes[0].text(x_pos,y_val,int(y_val),ha='center',va='bottom')
    bar.set_edgecolor('white')
    bar.set_hatch('/')
  axes[0].grid(axis='y',linestyle='--',linewidth=0.7)
  axes[0].set_axisbelow(True)
  sns.despine(left=True,bottom=True)
  # Pie (Donut) Chart for the distribution percentages
  axes[1].pie(df[col_name].value_counts().values,
              labels=df[col_name].value_counts().index,
              colors=['#7ecfed','#eda17e'],
              autopct='%1.2f%%',
              startangle=90)
  axes[1].set_title(f'Percentages of {col_name} Distribution')
  center_circle3 = plt.Circle((0,0),radius=0.8,fc='white')
  fig = plt.gcf()
  fig.gca().add_artist(center_circle3)
  plt.tight_layout()
  plt.show()

for col in categorical_columns:
  Distribution_of_Cat_cols(col,df)
```

```python
# Average infection of Lung cancer per gender based on possible symptoms
df[categorical_columns].head()

def Average_infection(col_name,data):
  #average = data.groupby(col_name)['LUNG_CANCER'].mean()
  plt.figure(figsize=(7,5),dpi=100)
  bars = sns.barplot(x=col_name,
                     y='LUNG_CANCER',
                     data=df,
                     estimator='mean',
                     palette=custom_palette,
                     errorbar=None,
                     hue='GENDER',
                     dodge=True)
  for bar in bars.patches:
    y_val = bar.get_height()
    x_pos = bar.get_x() + (bar.get_width()/2)
    if y_val > 0:
      plt.text(x_pos,y_val,f'{round((y_val*100),2)}%',ha='center',va='bottom')
    bar.set_edgecolor('white')
    bar.set_hatch('/')
  bars.grid(axis='y',linestyle='--',linewidth=0.7,color='gray')
  bars.set_axisbelow(True)
  sns.despine()
  plt.title(f'Average Infection per {col_name}',fontsize=12)
  plt.ylabel('Average Infection')
  plt.xlabel(col_name)
  plt.tight_layout()
  plt.savefig(f'Average Infection per {col_name}.png')
  plt.show()
  print('\n')
for col in categorical_columns:
  Average_infection(col,df)
```
